## Pyplot

Pyplot is a module in Matplotlib that provides a MATLAB-like interface to create plots. It is a collection of functions that make matplotlib work like MATLAB. 

Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias:



```python
import matplotlib.pyplot as plt
```

Matplotlib graphs your data on Figures (e.g., windows, Jupyter widgets, etc.), each of which can contain one or more Axes, an area where points can be specified in terms of x-y coordinates. The simplest way of creating a Figure with an Axes is using pyplot.subplots. We can then use Axes.plot to draw some data on the Axes:

```python
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
```

**Marker Size**

## Title and Labels

With Pyplot, you can use the **xlabel()** and **ylabel()** functions to set a label for the x- and y-axis. 

And you can use the **title()** function to set a title for the plot.

```python
x = np.array([0,2,5,10 ])
y = np.array([3,7,5, 10])
plt.plot(x, y)

plt.title("My Title")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.show()
```

### You can also change fonts

use the fontdict parameter in xlabel(), ylabel(), and title() to set font properties for the title and labels.

```python
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.title("My Title", font1)
plt.xlabel("X-Axis",font2)
plt.ylabel("Y-Axis",font2)
plt.plot(x, y)
plt.show()
```

### we can also position title

loc parameter in title() is used to position the title.

```python
plt.title("My Title", font1,loc='center')
plt.xlabel("X-Axis",font2,loc='right')
plt.ylabel("Y-Axis",font2)
plt.plot(x, y)
plt.show()
```

## Gridlines

**grid()** function can be used to add grid lines to the plot.

```python
plt.title("My Title", font1,loc='left')
plt.xlabel("X-Axis",font2)
plt.ylabel("Y-Axis",font2)
plt.plot(x, y)
plt.grid()
plt.show()
```

#### specifying the gridline

```python
plt.title("My Title", font1,loc='left')
plt.xlabel("X-Axis",font2)
plt.ylabel("Y-Axis",font2)
plt.plot(x, y)
plt.grid(axis = 'x')
plt.show()
```

**Note:** we can also set line property to gridline

```python
plt.title("My Title", font1,loc='left')
plt.xlabel("X-Axis",font2)
plt.ylabel("Y-Axis",font2)
plt.plot(x, y)
plt.grid(axis = 'x',linestyle='--',lw=0.3,color='g')
plt.show()
```

## Subplot

Multiple plots on single figure

The subplot() function in Matplotlib is used to create multiple plots on the same figure. It takes three arguments:
* nrows: : The number of rows of subplots.
* ncols: : The number of columns of subplots.
* index: : The index of the current subplot.

The index starts at 1 and increases from left to right, then from top to bottom.

```python

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 2, 2)
plt.plot(x,y)

#plot 3:
x = np.array([1, 5, 12, 13])
y = np.array([3, 10, 4, 5])

plt.subplot(2, 2, 3)
plt.plot(x,y)

#plot 4:
x = np.array([1, 1, 4, 9])
y = np.array([5, 2, 11, 20])

plt.subplot(2, 2, 4)
plt.plot(x,y)
plt.show()
```

**We can add title to each plot**

```python

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 2, 1)
plt.plot(x,y)
plt.title("Plot1")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 2, 2)
plt.plot(x,y)
plt.title("Plot2")

#plot 3:
x = np.array([1, 5, 12, 13])
y = np.array([3, 10, 4, 5])

plt.subplot(2, 2, 3)
plt.plot(x,y)
plt.title("Plot3")

#plot 4:
x = np.array([1, 1, 4, 9])
y = np.array([5, 2, 11, 20])

plt.subplot(2, 2, 4)
plt.plot(x,y)
plt.title("Plot4")

plt.show()
```

### Super Title

You can add a title to the entire figure with the suptitle() function:

```python

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 2, 1)
plt.plot(x,y)
plt.title("Plot1")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 2, 2)
plt.plot(x,y)
plt.title("Plot2")

#plot 3:
x = np.array([1, 5, 12, 13])
y = np.array([3, 10, 4, 5])

plt.subplot(2, 2, 3)
plt.plot(x,y)
plt.title("Plot3")

#plot 4:
x = np.array([1, 1, 4, 9])
y = np.array([5, 2, 11, 20])

plt.subplot(2, 2, 4)
plt.plot(x,y)
plt.title("Plot4")

plt.suptitle("SuperTitle")
plt.show()
```

## Histogram

A histogram is a graphing tool that displays the distribution of data points across a range of numerical values. It's a column chart that shows frequency data. 


**hist()** function is used to create histograms.

The hist() function will use an array of numbers to create a histogram, the array is sent into the function as an argument.

```python
x = np.random.normal(170,10,100)
plt.hist(x)
plt.show()
```

```python
x = np.random.normal(170,10,100)
plt.hist(x,color='r',width=1,alpha=0.5)
plt.show()
```

## Pie Chart

pie() function is used to draw pie charts:

```python

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show() 
```

### labels on the chart


Add labels to the pie chart with the labels parameter.

The labels parameter must be an array with one label for each wedge:

```python
mylabels = ["team a", "team b", "team c", "team d"]

y = np.array([35, 25, 25, 15])

plt.pie(y, labels =mylabels)
plt.show() 
```

We can also set start anlge useing startangle parameter

```python

plt.pie(y, labels =mylabels,startangle = 180)
plt.show() 
```

**Explode**

Maybe you want one of the wedges to stand out? The explode parameter allows you to do that.

The explode parameter, if specified, and not None, must be an array with one value for each wedge.

Each value represents how far from the center each wedge is displayed:

```python
Explode = [0.1, 0.1, 0.1, 0.1]
```

```python
plt.pie(y, labels =mylabels,startangle = 90, explode=Explode)
plt.show() 
```

### Colors


You can set the color of each wedge with the colors parameter.

The colors parameter, if specified, must be an array with one value for each wedge:

```python
Color_list=['Red','Blue','Green','Yellow']
plt.pie(y, labels =mylabels,startangle = 90, explode=Explode,colors=Color_list)
plt.show() 
```

### Legend

To add a list of explanation for each wedge, use the legend() function:

```python
plt.pie(y)
plt.legend(mylabels)
plt.show() 
```

**We can also add title to legend**

```python
plt.pie(y)
plt.legend(mylabels, title="MyLegend")
plt.show() 
```

**We can also set position of legend**

```python
Color_list=['Red','Blue','Green','Yellow']
plt.pie(y)
plt.legend(mylabels, title="MyLegend", loc='lower center')
plt.show() 
```
